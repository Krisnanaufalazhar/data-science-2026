```python
import pandas as pd
import numpy as np
```

## Load Dataset

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df.head()
```

## Q1: Jumlah baris, kolom, dan nama kolom

```python
df.shape, df.columns
```

## Q2: Missing values

```python
missing = df.isnull().sum()
persen = (missing / len(df)) * 100
pd.DataFrame({'Missing': missing, 'Persen (%)': persen})
```

## Q3: Persentase penumpang selamat

```python
df['Survived'].mean() * 100
```

## Q4: Wanita kelas 1

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
wanita_kelas1.shape[0], wanita_kelas1['Age'].mean()
```

## Q5: Survival rate per kelas

```python
df.groupby('Pclass')['Survived'].mean()
```

## Q6: Analisis tambahan (Gender)

```python
df.groupby('Sex')['Survived'].mean()
```
